```python
import pandas as pd
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
dataset=pd.read_csv('Facies Data.csv')
x=dataset.iloc[:,4:11] 
y=dataset.iloc[:,0].values
import numpy as np
seed=50 
np.random.seed(seed)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y,
 test_size=0.3)
model = make_pipeline(MinMaxScaler(),RandomForestClassifier(max_features=2,
                                                         n_estimators=80,
                                                         random_state=0),)
rfmodel1=model.fit(x_train,y_train)
from joblib import dump
dump(rfmodel1,"rfmodel1.joblib")
```

```python
import pandas as pd
from joblib import load
dataset=pd.read_csv('Facies Data.csv')
x=dataset.iloc[:,4:11]
y=dataset.iloc[:,0].values
import numpy as np 
seed=50
np.random.seed(seed)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y,
test_size=0.3)
Model=load('rfmodel1.joblib')
yModel=Model.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, yModel))
```
